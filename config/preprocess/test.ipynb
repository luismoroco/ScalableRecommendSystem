```python
import pandas as pd
```

```python
df_movies = pd.read_csv('..//../data/ml-latest/movies.csv')
df_movies.head(10)
```

```python
df_movies.isna().sum()
```

```python
df_movies.drop_duplicates()
```

```python
df_movies['genres'] = df_movies['genres'].str.split('|')
unique_genres = []
for genres in df_movies['genres']:
    unique_genres.extend(genres)

unique_genres = set(unique_genres)
print(unique_genres)
```

```python
genre_dict = {genre: id+1 for id, genre in enumerate(unique_genres)}
```

```python
genre_dict
```

```python
genre_name = "Adventure"
genre_id = genre_dict.get(genre_name)
print(genre_id)
```

```python
genres_with_id = [(i + 1, genre) for i, genre in enumerate(unique_genres)]
df_genres = pd.DataFrame(genres_with_id, columns=['genre_id', 'genre'])
df_genres.head(10)
```

```python
df_genres.to_csv('genres.txt', sep='\t', encoding='utf-8', header=False, index_label=False, index=False)
```
